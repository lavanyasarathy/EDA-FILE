1. Customer Analysis

#Demographic Distribution

```python
# Calculate age from birthday
df['Age'] = (pd.Timestamp.now() - df['Birthday']).dt.days // 365

# Demographic Distribution
gender_distribution = df['Gender'].value_counts()
age_distribution = df['Age'].value_counts(bins=10)  # Create bins for age
location_distribution = df['City'].value_counts().head(10)  # Top 10 cities

# Plotting
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
gender_distribution.plot(kind='bar', title='Gender Distribution')
plt.subplot(1, 3, 2)
age_distribution.plot(kind='bar', title='Age Distribution')
plt.subplot(1, 3, 3)
location_distribution.plot(kind='bar', title='Top 10 Cities')
plt.tight_layout()
plt.show()
```

#Purchase Patterns

```python
# Average Order Value
avg_order_value = df['Unit Price USD'].mean()
# Purchase Frequency
purchase_frequency = df['Order Number'].nunique() / df['CustomerKey'].nunique()
# Preferred Products
preferred_products = df['Product Name'].value_counts().head(5)

# Print results
print(f"Average Order Value: {avg_order_value}")
print(f"Purchase Frequency: {purchase_frequency}")
print("Preferred Products:\n", preferred_products)
```

 **2.Sales Analysis**


#Overall Sales Performance

```python
# Total Sales Over Time
df['Sales'] = df['Quantity'] * df['Unit Price USD']
sales_over_time = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Plotting
sales_over_time.plot(kind='line', title='Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

#Sales by Product

```python
sales_by_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

# Plotting
sales_by_product.head(10).plot(kind='bar', title='Top 5 Products by Sales')
plt.ylabel('Sales')
plt.show()
```

#Sales by Store

```python
sales_by_store = df.groupby('StoreKey')['Sales'].sum()

# Plotting
sales_by_store.plot(kind='bar', title='Sales by Store')
plt.ylabel('Sales')
plt.show()
```

3. Product Analysis

#Product Popularity

```python
product_popularity = df['Product Name'].value_counts()

# Plotting
product_popularity.head(10).plot(kind='bar', title='Most Popular Products')
plt.ylabel('Number of Sales')
plt.show()
```

Profitability Analysis

```python
df['Profit'] = df['Unit Price USD'] - df['Unit Cost USD']
profitability = df.groupby('Product Name')['Profit'].mean()

# Plotting
profitability.sort_values(ascending=False).head(10).plot(kind='bar', title='Top 10 Profitable Products')
plt.ylabel('Average Profit')
plt.show()
```

4. Store Analysis

#Store Performance

```python
store_performance = df.groupby('StoreKey')['Sales'].sum()

# Plotting
store_performance.plot(kind='bar', title='Store Performance')
plt.ylabel('Sales')
plt.show()
```

#Geographical Analysis

```python
# Assuming you want to analyze sales by Country
sales_by_country = df.groupby('Country')['Sales'].sum().sort_values(ascending=False)

# Plotting
sales_by_country.head(10).plot(kind='bar', title='Sales by Country')
plt.ylabel('Sales')
plt.show()
```
